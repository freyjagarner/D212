## Part II: Market Basket Justification

### B.  Explain the reasons for using market basket analysis by doing the following:

#### 1.  Explain how market basket analyzes the selected dataset. Include expected outcomes.

Market basket analysis looks at the frequency of items being purchased together to try to identify a relationship between the items. 

The expected outcome of this analysis will be:
* The **antecedent** which is the initial purchase
* The **consequent** which is the item purchased as a result of the first purchase
* The **support** which is the ratio of transactions which contain an item or items.
* The **confidence** which is the ratio of the support of an item or items to the support of an item or items.
* The **lift** which is the ratio of the transactions containing an item or items to the item or items if they were random and independent, or the ration of the true to expected confidence.

```python
# import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.preprocessing import TransactionEncoder
```

```python
# import data from csv
df = pd.read_csv('teleco_market_basket.csv')

# set it so we can see all columns
pd.set_option('display.max_columns', None)
```

```python
trans_list = df.stack().groupby(level=0).apply(list).tolist()
```

```python
encoder = TransactionEncoder().fit(trans_list)
oh = encoder.transform(trans_list)
oh_df = pd.DataFrame(oh, columns = encoder.columns_)
oh_df
```

```python
def find_items(df, name_list):
    item_list = []
    for i in df.columns:
        for x in name_list:
            if i.lower().find(x) >= 0:
                item_list.append(i)
    return item_list
```

```python
ipc = ['iphone charger', 'lightning cable', 'lighning cable', 'iphone 12 charger cable', 'lightning to usb']
adapters = ['adapter']
c_cables = [' c cable', 'c charger', 'usb type c to usb']
bl_glasses = ['blue light']
comp_gas = ['compressed gas']
eth_cable = ['ethernet cable']
sdc = ['gb card', 'microsdxc', 'sdhc', 'sandisk']
pi = [' ink', 'toner']
headsets = ['headset']
modems = ['modem']
routers = ['router', 'home mesh']
ip_case = ['case']
mice = [' mouse']
app_pen = ['apple pencil', 'stylus']
hubs = ['hub']
wcs = ['webcam']

iphone_charger = find_items(oh_df, ipc)
adapter = find_items(oh_df, adapters)
type_c = find_items(oh_df, c_cables)
bluelight_glasses = find_items(oh_df, bl_glasses)
compressed_gas = find_items(oh_df, comp_gas)
ethernet_cable = find_items(oh_df, eth_cable)
sd_card = find_items(oh_df, sdc)
printer_ink = find_items(oh_df, hpi)
headset = find_items(oh_df, headsets)
modem = find_items(oh_df, modems)
router = find_items(oh_df, routers)
iphone_case = find_items(oh_df, ip_case)
mouse = find_items(oh_df, mice)
apple_pencil = find_items(oh_df, app_pen)
usb_hub = find_items(oh_df, hubs)
webcam = find_items(oh_df, wcs)

new_cols = iphone_charger + type_c + adapter+ bluelight_glasses + compressed_gas + ethernet_cable + sd_card + hp_ink + headset + modem + router + iphone_case + mouse + apple_pencil + usb_hub + webcam
```
